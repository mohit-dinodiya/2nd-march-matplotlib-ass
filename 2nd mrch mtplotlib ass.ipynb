# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot or graph that displays the relationship between two variables in a dataset. It is a two-dimensional plot that uses dots or markers to represent individual data points. The position of each dot on the plot represents the values of the two variables for that data point.

Scatter plots are often used to identify patterns or relationships between two variables. For example, if the two variables are positively correlated, meaning that as one variable increases, the other variable tends to increase as well, the dots on the scatter plot will tend to cluster in a line with a positive slope. Similarly, if the two variables are negatively correlated, meaning that as one variable increases, the other variable tends to decrease, the dots on the scatter plot will tend to cluster in a line with a negative slope.

Scatter plots can also be used to identify outliers, or individual data points that fall far away from the general pattern of the data. Outliers can be important to identify, as they can have a significant impact on the relationship between the two variables.

Overall, scatter plots are a useful tool for visualizing the relationship between two variables in a dataset, and can be used in a wide range of fields, including statistics, data science, and social sciences.

```python
import numpy as np
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("title:x vs y")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots or subplots within a single figure. It allows us to create a grid of plots with a specified number of rows and columns, and then specify which subplot we want to plot on using indexing.

For example, if we want to create a figure with two subplots side by side, we can call subplot(1,2,1) to create the first subplot in the first column, and then subplot(1,2,2) to create the second subplot in the second column.

Using subplot() can be useful when we want to compare different visualizations of the same data or plot related data in a compact and organized manner. Additionally, using subplots can save space in a visualization and make it easier to compare and contrast different aspects of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x1,x2)
axes[0,1].plot(x2,y2)
axes[1,0].plot(x3,y3)
axes[1,1].plot(x4,y4)


fig.suptitle('four line plots')

plt.show()



```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart, is a way of representing data using rectangular bars. The length of each bar corresponds to the value of a particular variable, and the bars can be arranged horizontally or vertically.

Bar plots are commonly used to compare the values of different variables or groups, and to visualize patterns or trends in data. They are particularly useful when dealing with categorical data, as they allow us to easily compare the values of different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.title('company"s profit')
plt.xlabel('company')
plt.ylabel('profit')
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit)
plt.title('company"s profit')
plt.xlabel('company')
plt.ylabel('profit')
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a way of summarizing the distribution of a set of continuous data. The box in the plot represents the middle 50% of the data, with the median indicated as a horizontal line inside the box. The "whiskers" extend from the box to the minimum and maximum values that fall within 1.5 times the interquartile range (IQR) of the data, and any values outside this range are shown as individual points.

Box plots are commonly used to compare the distributions of different variables or groups, and to identify outliers or extreme values in the data. They provide a quick and easy way to visualize the central tendency, spread, and shape of a dataset

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100,10,200)

box2 = np.random.normal(90,20,200)

data = [box1,box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Values')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()

```
